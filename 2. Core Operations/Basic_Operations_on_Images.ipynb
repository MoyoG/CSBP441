```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

Accessing and Modifying pixel values

```python
img = cv.imread('../images/messi5.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
```

```python
px = img[100,100]
print("The color code at the 100x100 coordinate is:  ",px)
#accessing only blue pixel
blue = img[100,100,0]
print("The blue color value at the 100x100 coordinate is:", blue)
```

```python
#You can modify the pixel values the same way.
img[100,100] = [255,255,255]
print(img[100,100])
```

```python
#Better pixel accessing and editing method :
# accessing RED value
red_px = img.item(10,10,2)
print("The color code at the 10x10 coordinate is:",red_px)
# modifying RED value
img.itemset((10,10,2),100)
change_px = img.item(10,10,2)
print("The color code at the 10x10 coordinate is:",change_px)
```

Accessing Image Properties

```python
print('image size : ',img.size )
print('image data type : ',img.dtype )
```

Image ROI

```python
ball = img[330:390, 390:460] 
img[330:390, 100:170] = ball 
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.axis('off')  # Hide the axis
plt.show()
```

Splitting and Merging Image Channels

```python
b,g,r = cv.split(img)
img = cv.merge((b,g,r))
#Or 
b = img[:,:,0]
img[:,:,2] = 0
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.axis('off')  # Hide the axis
plt.show()
```

Making Borders for Images (Padding)

```python
border_color = [255,255,0]
img1 = cv.imread('../images/opencv_logo.png')
print(img1.shape)
assert img1 is not None, "file could not be read, check with os.path.exists()"
replicate = cv.copyMakeBorder(img1,40,40,40,40,cv.BORDER_REPLICATE)
reflect = cv.copyMakeBorder(img1,40,40,40,40,cv.BORDER_REFLECT)
reflect101 = cv.copyMakeBorder(img1,40,40,40,40,cv.BORDER_REFLECT_101)
wrap = cv.copyMakeBorder(img1,40,40,40,40,cv.BORDER_WRAP)
constant= cv.copyMakeBorder(img1,100,100,100,100,cv.BORDER_CONSTANT,value=border_color)
plt.subplot(231),plt.axis('off'),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.axis('off'),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.axis('off'),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.axis('off'),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.axis('off'),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.axis('off'),plt.imshow(constant,'gray'),plt.title('CONSTANT')
plt.show()
```
