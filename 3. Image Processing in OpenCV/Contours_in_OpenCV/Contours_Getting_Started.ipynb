Import libraries

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
im = cv.imread('../../images/box.jpg')
assert im is not None, "File could not be read, check with os.path.exists()"

imgray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(imgray, 127, 255, 0)
# Find contours (ignoring hierarchy)
contours, _ = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
# Draw contours on the original image
result = im.copy()
cv.drawContours(result, contours, -1, (0, 255, 0), 2)

plt.subplot(121),plt.imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB)),plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB)),plt.title('Found contour'), plt.xticks([]), plt.yticks([])
plt.show()
```
