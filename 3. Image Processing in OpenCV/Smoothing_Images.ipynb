Import libraries

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

2D Convolution ( Image Filtering )


```python
img = cv.imread('../images/opencv_black.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
kernel = np.ones((5,5),np.float32)/25
dst = cv.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(cv.cvtColor(dst, cv.COLOR_BGR2RGB)),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

Gaussian Blurring

```python
blur = cv.GaussianBlur(img,(5,5),0)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

Median Blurring

```python
img_noisy = cv.imread('../images/opencv_noisy.jpg')
median = cv.medianBlur(img_noisy,3)
plt.subplot(121),plt.imshow(cv.cvtColor(img_noisy, cv.COLOR_BGR2RGB)),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(cv.cvtColor(median, cv.COLOR_BGR2RGB)),plt.title('Median')
plt.xticks([]), plt.yticks([])
plt.show()
```

Bilateral Filtering

```python
img1 = cv.imread('../images/bilateral.jpg')
img1_bilateral = cv.bilateralFilter(img1,9,75,75)
plt.subplot(121),plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB)),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(cv.cvtColor(img1_bilateral, cv.COLOR_BGR2RGB)),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
